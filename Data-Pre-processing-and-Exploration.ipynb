<h2 id = "1">Task 1</h2> 
<h3>Introduction:</h3>
<p><b style = "color: red">Stroke</b> is a severe medical condition that arises due to the improper or no flow of oxygen and nutrients by the arteries to the brain of a person which further affects the functionality of the body. This medical issue is known widely across the globe. Moreover, according to <i>World Health Organization</i> (WHO) it is the 2nd leading cause of death in all the nations. It is very common and can happen to anyone also recovery from a stroke can be distressing and devastating as it is a slow process. The after effects of a stroke usually involves loss of vision, paralysis, communication problems, emotional and physical changes. In addition to this, there are several lifestyle factors such as smoking, hypertension, stress, etc, which have a serious impact on the health of the person. Furthermore, there are different types of stroke, where for some people the effect might be mild and for others it can be severe with disability continued forever. Henceforth, to mitigate such health issue we aim to predict the likelihood of a person to experience a stroke such that some precautionary health measures could be taken in order to avoid such life-threating health issue.</p>

<h3>Two Variable Visaulisations</h3>

```python
print('\033[1m' +'\033[98m' + "\n \nTwo Variable Visualizations:")
print('\033[1m' +'\033[91m' + "\nNumber of Males and Females with Hypertension:")
sns.countplot(x='hypertension', hue='gender', data=df, palette='spring');
plt.xlabel('Hypertension')
plt.ylabel('Count')
plt.title('Number of Males and Females with Hypertension')
plt.show()
print('\033[1m' +'\033[98m' + "From the above bar chart, it is evident the women have more hypertension than men. The graph also shows that there are more than 2500 women and above 1500 men who didn't have hypertension.")

# Stroke in various Age Groups
print('\033[1m' +'\033[91m' + "\nStroke in various Age Groups:")
plot2 = sns.kdeplot(df['age'][(df["stroke"] == 0)] , color="Red", shade = True)
plot2 = sns.kdeplot(df['age'][(df["stroke"] == 1)], ax =plot2, color="Blue", shade= True)
plot2= plot2.legend(["No,","Yes"],fontsize = 11)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Stroke in various Age Groups')
plt.show()
print('\033[1m' +'\033[98m' + "The density plot above displays the different age groups that experience stroke. After plotting this visualisation, it was discovered that the age of people ranging from 40 to 90 are more likeable to get a stroke(by observing the blue line), whereas the frequency of people not getting a stroke reduces in this age period(as indicated by the red line in the graph.)")

# Smoking Status By Gender
print('\033[1m' +'\033[91m' + "\nSmoking Status By Gender:")
sns.countplot(x='smoking_status', hue='gender', data=df)
plt.xlabel('Smoking Staus')
plt.ylabel('Count')
plt.title('Smoking Status By Gender')
plt.show()
print('\033[1m' +'\033[98m' + "With the help of the above illustration, the smoking staus for the gender can be observed. It can be seen from the bar chart that the number of females who never smoked is much more than the number of males. Also for both categories i.e. 'formerly smoked' and 'currently smokes', there were more women as compared to men. ")

# Average Glucose Level by Heart Disease
print('\033[1m' +'\033[91m' + "\nAverage Glucose Level by Heart Disease:")
sns.violinplot( x = df["heart_disease"],y = df["avg_glucose_level"], palette = "Set1")
plt.xlabel('Heart Disease')
plt.ylabel('Average Glucose Level')
plt.title('Average Glucose Level by Heart Disease')
plt.show()
print('\033[1m' +'\033[98m' + "The violin plot compares the average glucose level present in the body for the group of people who have and don't have heart disease. The observed average glucose level for people who didnt have heart disease was between 50 to 125 range and was more as compared to people with heart disease. Moreover, as the average glucose level increased, people suffering from a heart disease also increased.")
```

<h3>Three Variable Visualizations</h3>

```python
print('\033[1m' +'\033[98m' + "\n \nThree Variable Visualization")

df["Hypertension"] = df["hypertension"].map(dict(zip(range(0,2), ["No", "Yes"])))
df["Heart Disease"] = df["heart_disease"].map(dict(zip(range(0,2), ["No", "Yes"])))

# Smoking status vs age vs heart disease
print('\033[1m' +'\033[91m' + "\n Smoking Status, Age, Heart Disease:")
viz1 = sns.lmplot(x='smoking_status', y='age', hue='heart_disease', markers=['x', 'o'], fit_reg=False, data=df, legend = False);
plt.legend(title='Heart Disease', loc='upper left', labels=['No', 'Yes'], bbox_to_anchor=(1.2, 1))
plt.xlabel('Smoking Status')
plt.ylabel("AGE")
plt.title('Effect of Smoking with Age on Heart')
plt.show(viz1)
print('\033[1m' +'\033[98m' + "From the above visualization we can clearly see that as the age increases the chances of having a heart disease also increases and its higher in people who have smoked earlier or smokes.")


# age vs glucose level vs heart disease 
print('\033[1m' +'\033[91m' + "\n Age, Glucose Level, Heart Disease:")
viz2 = sns.lmplot(x='age', y='avg_glucose_level', hue='heart_disease', 
           markers=['+', 'o'],
           fit_reg=False, data=df, legend = False);
plt.legend(title='Heart Disease', loc='upper left', labels=['No', 'Yes'], bbox_to_anchor=(1.2, 1))
plt.xlabel("AGE")
plt.ylabel('Average Glucose Level')
plt.title('Effect of Low/high Glucose Level on Heart with Age')
plt.show(viz2)
print('\033[1m' +'\033[98m' + "We can see from the scatter plot that younger peole have comparatively lower glucose level as compared to older people and the risk of having heart disease is higher in the latter than the former. Also as the age increases the glucose level also increases and hence the risk of heart disease also increases.")


# Heart disease vs age vs hypertension
print('\033[1m' +'\033[91m' + "\n Heart Disease, Age, Hypertension:")
pl = sns.barplot(x="Heart Disease", y="age", hue='Hypertension', data=df)
pl.legend(title="Hypertension", loc=9, bbox_to_anchor=(1.2, 1));
plt.xlabel('Suffering from Heart Disease')
plt.ylabel("AGE")
plt.title('Effect of Hypertension on Heart with Age')
plt.show(pl)
print('\033[1m' +'\033[98m' + "It can be observed from the bar chart that people of older age with heart disease have a higher probability of having hypertension when compared to younger people.")

# BMI, Age, Stroke
print('\033[1m' +'\033[91m' + "\n BMI, Age, Stroke:")
bas = sns.scatterplot(x="bmi", y="age", data=df[df['stroke'] == 0], alpha=0.5)
sns.scatterplot(x="bmi", y="age",ax=bas, alpha=0.8, data=df[df['stroke'] == 1])
plt.legend(title='Chances of getting a Stroke', loc='upper left', labels=['No', 'Yes'], bbox_to_anchor=(1.2, 1))
plt.xlabel('Age')
plt.ylabel("BMI (Body Mass Index)")
plt.title('Chances of getting a stroke due to Low/high BMI w.r.t Age')
plt.show(bas)
print('\033[1m' +'\033[98m' + "From this visualization it is pretty evident that people with BMI more than 25 have severe chances of getting a stroke, moreover, it can also be noticed that, people with age 40 or greater have high chances of getting a stroke and once reaching 60 it increases even more.")
```
